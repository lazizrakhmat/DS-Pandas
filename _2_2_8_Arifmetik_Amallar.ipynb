ARIFMETIK AMALLAR

Ikki ```Series``` yoki ```DataFrame``` obyektlari o'rtasida arifmetik amallar bajarishda indexlar juda muhim rol o'ynaydi.

Amallar mos tushuvchi indexlar o'rtasida bajariladi. Agar obyektlarning birida mavjud indeks ikkinchisida yo'q bo'lsa NaN bo'ladi.



`````python
import pandas as pd
import numpy as np
`````

`````python
s1 = pd.Series([7.3, 4.5, -2.3, 1.6], index=['a', 'b', 'c', 'd']) 
s2 = pd.Series([-2.1, 3.6, 1.3, 4.8, 2], index=['a', 'c', 'e', 'f', 'g'])
`````

`````python
s1
`````

`````python
s2
`````

`````python
s1 + s2 
`````

E'tibor bering, ````s1```` da mavjud ```d``` indeksi ```s2``` da bo'lmaganligi sabab ```s1+s2``` da ```d```  indeksidagi qiymat ```NaN```bo'ldi.

Yuqoridagi holat ```DataFrame```larga hos. ```Series```lardan farqli ravshda ```DataFrame``` da qator va ustun indekslari hisobga olinadi.

`````python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma', 'Anor', 'Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Qovun', 'Anjir', 'Gilos', 'Olma'])
`````

`````python
df1
`````

`````python
df2
`````

`````python
df1 + df2
`````

`````python
df1.add(df2, fill_value=0) # .add() 
`````

`````python
df2.sub(df1, fill_value=0) # .sub() - ayirish
`````

`````python
df2.rsub(df1, fill_value=0) # .rsub() - o'rnini almashtirib ayirish, ya'ni o'ng tomondan boshlab ayirish
`````

Bunda maxsus ```fill_value``` parametri yordamida mos tushuvchi indekslardagi NaN qiymatlarni boshqa qiymat bilan almashtirish mumkin. Bu usulda yakuniy natija ikkala df da ham mavjud bo'lmagan indekslarda NaN  qiymati qaytadi.

`````python
df2.div(df1, fill_value=0) # .div() - bo'lish 
`````

## **Boshqa arifmetik amallar**

*   add, radd - Qo'shish(+)
*   sub, rsub - Ayirish(-)
*   div, rdiv - Bo'lish(/)
*   floordiv, rfloordiv - Bo'lish va butun qismini olish(//)
*   mul, rmul - Ko'paytirish(*)
*   pow, rpow - Eksponenta(**)

Yuqoridagi amallarning barchasida ```fill_value``` parametri mavjud.

```r```  bilan boshlangan methodlarda ```DataFrama``` ifodaning o'ng qiymatida bo'ladi.
