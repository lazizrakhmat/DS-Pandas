```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        "Zichlik":[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}
df = DataFrame(data)
```

```python
df
```

```python
df.index = [21, 20, 19, 18, 17, 16, 15, 10]
```

```python
df
```

```python
# Reindex
ind = [21, 20, 19, 18, 17, 16, 15, 10]
ind.sort()
ind
```

```python
df = df.reindex(ind)
```

```python
df
```

```python
df.reindex(index=range(10,22), method='nearest')
```



*  `nearest` bu method eng yaqin qiymatlarni oladi va bo'sh qolib ketgan qiymatlarni to'ldiradi

*   `ffill` forward fill bu esa eng oxirgi qiymatlarni olib bo'sh qolib ketgan qiymatlar bilan to'ldiradi



```python
df.reindex(index=range(10, 22), method='ffill')
df
```
