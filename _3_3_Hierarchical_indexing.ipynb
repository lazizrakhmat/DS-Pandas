```python
df = pd.read_csv("/content/Practicum_Datasets/world_cities_air_water_quality.csv")
```

```python
df
```

```python
df.info()
```

```python
columns = ['City','Region','Country','AirQuality','WaterPollution'] # Column -larni ("") larnsiz saqlab oldik
df.columns = columns
```

```python
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"',"")) # Bu yerda ("") olib tashladik
df['Region'] = df['Region'].map(lambda x: x.replace('"',"")) 
```

```python
df[df['Country']=='Germany'] # ma'lumotlar qaytmaydi 
```

```python
df['Country'][2] # bu yerda _United States of America ya'ni boshida space borligi ma'lum bo'ldi
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip()) # bu yerda biz lambda va strip  funksiyasidan foyldalanib space'ni olib tashladik
```

```python
df[df['Country']=='Germany']
```

```python
df.to_csv("Updated_dataset.csv", index=False) # saqlab oldik
```
