```python
import pandas as pd
import numpy as np
```

Pandas Obyeki elementlariga Funksiyalar qo'llash

```python
df = pd.DataFrame(np.random.randn(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
df
```

```python
np.abs(df) # .abs() - bu function bizga manfiy bo'lsaham musbas qiymat qaytaradi
```

```python
np.round(df) # .round() - bu function bizga butun qiymat qaytaradi
```

```python
df 
```

```python
df.mean() # .mean() bu function bizga o'rtacha qiymatni qaytaradi, ustunlari bo'yicha
```

```python
df.mean(axis=1) # .mean(axis=1) bu esa qatorlari bo'yicha
```

```python
df.apply(np.abs)
```

```python
df.apply(sum, axis=1)
```

```python
f = lambda x: x.max() - x.min() # lambda bizda bu yerda x.max() va x.min() qiymatlar o'rtachasini topib berdi
```

```python
df.apply(f) 
```
