# Qiymatlarni o'zgartirish
  

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv')
df.head()
```

```python
df.duplicated() # takrorlangan qatorlarni ko'rsatadi va olib tashlaydigan function - TRUE qaytaradi
```

```python
df.shape # df - dagi ma'lumotlarni hajmini ko'rib oldik
```

```python
df.duplicated().sum() # df - dagi duplicated bo'lgan qatorlarimizni yig'indisini olamiz
```

` drop_duplicates ` - bu function duplicated bo'lgan qatorlarni olib tashlaydi, agar hamma row'dagi ma'lumotlar bir xil bo'lsa, natijadan keyin indexlar o'zgaradi.

```python
df.drop_duplicates() # drop_duplicates() - function agar biz direct berib yo'naltirmasak unda uzi aynan hamma row larni faqat bir xil bo'lganlarini olib tashlaydi
```

```python
df.drop_duplicates(['country']) # biz bu yerda yo'naltirdik, ya'niki faqat 'country' dagi nom bir xil bo'ldimi olib tashla degan bo'yroq
```

#.map() - qiymatlarni moslash (almashtirish)

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df['num-of-cylinders'].unique() # unique - qiymatlarini ko'rib oldik 
```

```python
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12} # dict - tuzib olamiz sababi matnni songa o'zgartirishimiz uchun
```

```python
df['num-of-cylinders'].map(dict) # .map - function orqali biz moslashtirib oldik dict'ni ustunga
df['cylinders'] = df['num-of-cylinders'].map(dict) # yangi ustun nomiga save qilib olamiz
df
```

```python
df['price'] # ustundagi narxlar bizga $ ko'rinishida  
```

```python
kurs = 10720 # hozirgi kurs qiymatini berdik
df['price'].map(lambda x:x*kurs) # .map va lambda orqali biz kursni uzbek so'mdagi valuta qiymatiga o'zgartirib oldik
df['price_uz'] = df['price'].map(lambda x:x*kurs) # yangi ustun nomiga saqlaymiz
df
```
