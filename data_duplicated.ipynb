# Qiymatlarni o'zgartirish
  

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv')
df.head()
```

```python
df.duplicated() # takrorlangan qatorlarni ko'rsatadi va olib tashlaydigan function - TRUE qaytaradi
```

```python
df.shape # df - dagi ma'lumotlarni hajmini ko'rib oldik
```

```python
df.duplicated().sum() # df - dagi duplicated bo'lgan qatorlarimizni yig'indisini olamiz
```

` drop_duplicates ` - bu function duplicated bo'lgan qatorlarni olib tashlaydi, agar hamma row'dagi ma'lumotlar bir xil bo'lsa, natijadan keyin indexlar o'zgaradi.

```python
df.drop_duplicates() # drop_duplicates() - function agar biz direct berib yo'naltirmasak unda uzi aynan hamma row larni faqat bir xil bo'lganlarini olib tashlaydi
```

```python
df.drop_duplicates(['country']) # biz bu yerda yo'naltirdik, ya'niki faqat 'country' dagi nom bir xil bo'ldimi olib tashla degan bo'yroq
```
