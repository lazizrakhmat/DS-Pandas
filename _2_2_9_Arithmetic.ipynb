### Arifmetik amallar

Ikki `Series` yoki `DataFrame` obyektlari o'rtasida arifmetik amallar bajarishda indekslar juda muhim rol o'ynaydi. 

Amallar mos tushuvchi indekslar o'rtasida bajariladi. Agar obyektlarning birida mavjud indeks ikkinchisda yo'q bo'lsa natija `NaN` bo'ladi.

## `DataFrame` va `Series` o'rtasida arifmetik amallar

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
```

```python
df
```

```python
np.abs(df) # .abs() - absolutely qiymat 
```

```python
np.round(df) # .round() - butunlash function
```

```python
df.mean(axis=1)
```

```python
df.apply(np.abs) # o'zimizdi functionni yozishimiz mumkin
```

```python
df.apply(sum, axis=1)
```

```python
f = lambda x: x.max() - x.min() # bir martalik functions
```

```python
df.apply(f)
```

```python
df.apply(lambda x: x.max() - x.min())
```

```python
def foo(x):
  return pd.Series([x.max(), x.min()], index=['max', 'min'])
```

```python
df.apply(foo)
```

```python
df
```

```python
df.applymap() # xar bitta element uchun qo'llab chiqadi
```

```python
df = pd.DataFrame(np.arange(1,13).reshape(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
```

```python
df
```

```python
df.applymap(lambda x: x*x) # bu yerda o'ziga o'zini ko'paytirdi
```
