```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv", index_col=0)
df.head()
```

```python
len(df) # df 'ning uzunligini tekshiramiz
```

```python
df.take(randlist) # .take - randlist'dan df'ga oladi
```

```python
df.sample(10) # .sample - 10 ta tasodifiy qiymatlarni qaytaradi
```

```python
df.sample(10, replace=True) # ,replace=True tasodiy qiymatlarni qaytaradi lekin ba'zilari takrorlanib keladi
```
