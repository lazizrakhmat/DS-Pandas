## `Tartiblash` (sort) va `Reytinglash` (ranking)

```python
obj = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
obj
```

```python
obj.sort_index() # .sort_index - bu faqat index'ga ko'ra tartiblaydi
```

```python
obj.sort_values() # .sort_values - bu function faqat qiymatlarga ko'ra tartiblaydi
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.shape
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1, inplace=True) # bu ustunlar bo'yicha tartiblaydi, (inplace=True) - saqlab qolish
```

```python
df
```

```python
df.sort_index(axis=1, ascending=False) # ascending - o'sish tartibida False, bunda teskarisida o'zgaradi
```

```python
df.sort_values(by='price', ascending=False) # by=price - narxlari bo'yicha tartiblash, ascending= bu o'sish tartbida emas, teskari
```

```python
df.sort_values(by=['price', 'mileage'], ascending=False)
```

```python
df.sort_values(by='year', ascending=False) 
```

```python
df.sort_values(by='condition')
```
