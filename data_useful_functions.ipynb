# Foydali funksiyalar

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv", index_col=0)
df.head()
```

```python
df.describe() # bu yerda ko'rganizdek age - mean, min, va max qiymatlar xaqiqatga yaqin emas
```

```python
df[df.age > 123].index # nega 123 age tarixda eng yuqori yosh bo'lgan. Biz u yoshdan yuqorilarini indexlarini ko'rib olamiz
```

```python
df[df.age < 0].index # va min qiymatlarda -85 qiymatni ko'rdik, endi biz 0 dan kichik qiymatlarniham indexlarini ko'rib olamiz
```

`apply` - function orqali biz uning ichida funksiya yozamiz va apply uni qo'llaydi

```python
df['age'] = df['age'].apply(np.abs) # abs - obsolyut qiymatlarni olamiz ya'ni age'dagi (-) qiymatlarni hammasini (+) ga o'zgartiramiz
```

```python
df[df.age < 0] # check qilamiz 0 dan kichik qiymatlar bor yoki yo'qligini
```

### *Katta qiymatlarni esa nima qilamiz?*


*   Agar biz uni tozalay olmasak, drop ya'ni tashlab yuboramiz
 



```python
idx = df[df.age > 123].index # saqlab oldik, va indexlarini olamiz
idx
```

```python
df.drop(index=idx, inplace=True) # idx - qiymatlarni tashlab yubordik, va inplace=TRUE orqali joyida saqlab olamiz
df
```

```python
df.describe()
```
