### 3.2 Ma'lumotlarni tozalash va tayyorlash
Data Science muhandisi sifatida sizning 80% vaqtingiz ma'lumotlarga ishlov berish bilan o'tadi. Jumladan, ma'lumotlarni o'qish, kamchiliklarni bartaraf etish, ma'lumotlarni shakl o'zgartirish, matnlar bilan ishlash va hokozo.
### Mavjud bo'lmagan qiymatlar bilan ishlash

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.isnull() # mavjud bo'lmagan qiymatlarga TRUE deb qaytaradi
```

```python
df.notnull() # yuqoridagini teskarinisi qaytaradi ya'ni FALSE
```

```python
df.isnull().sum() # bu True'larni umumiy soni nechtaligini hisoblaydi, ya'ni yig'inidisini aniqlaydi
```

```python
len(df)
```

```python
df.isnull().sum()/len(df)*100 # % da qatorda qancha bo'sh qiymat borligini ko'rsatadi
```

```python
df.dropna() # NaN qatorlarni tashlab yuborish
```

```python
len(df.dropna()) # NaN qiymatlar olib tashlangandan keyin qolgan soni
```

```python
df.dropna(axis=1) # ustunlar bo'yicha NaN qiymatlarni tashlab yuborish
```

```python
df.dropna(axis=1, thresh=48, inplace=True) # thresh=n ,chegara qo'yamiz, mavjud bo'lmagan qiymatlar, 48 tadan kam bo'lgan ustunlarni tashlab yuboramiz
```

```python
df2 = df.dropna(subset=['price', 'horsepower']) # subset - Agar ana shu ustunlarda NaN qiymatlar mavjud bo'lsa o'shaq qatorni tashlab yuborgin 
```
